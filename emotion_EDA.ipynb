## okt사용하여 토큰화 & soynlp사용하여 반복되는 문자 정제

```python
from konlpy.tag import Okt
from soynlp.normalizer import *
okt = Okt()
```

```python
#test shell
print(okt.morphs(data.loc[1,'Sentence'])) #okt 테스트

print(emoticon_normalize('ㅈ망겜 핵관리도 못하더니 서버관리도 못하냐? ㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋ', num_repeats=3)) #soynlp 테스트
print(okt.pos(emoticon_normalize('ㅈ망겜 핵관리도 못하더니 서버관리도 못하냐? ㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋ', num_repeats=3))) #soynlp 테스트
```

```python
okt_sentence = []
for i in range(len(data)):
  a = data.loc[i,'Sentence']
  emoticon_normalize(a,num_repeats=3)
  okt_sentence.append(okt.pos(a,norm=True,stem=True))
```

# 3. Word2Vec

```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors

text = []
for i in range(len(data)):
  sentence = okt.morphs(data.loc[i,'Sentence'], norm = True, stem = True) #norm은 정규화 여부, stem은 어근화 여부임.#정규화 normalization (입니닼ㅋㅋ -> 입니다 ㅋㅋ, 샤릉해 -> 사랑해)#어근화 stemming (입니다 -> 이다)
  text.append(sentence)                                                        
```

```python
text2 = [j for i in text for j in i]
```

```python
model = Word2Vec(sentences=[text2], size = 50, window = 2, min_count = 2, workers = 4, sg = 0)
model
```

```python
print(model.wv.most_similar(['언니']))
```
